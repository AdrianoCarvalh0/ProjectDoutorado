```python
import cv2
import numpy as np

def add_salt_pepper_noise(image, salt_prob=0.02, pepper_prob=0.02):
    """
    Adiciona ruído sal e pimenta a uma imagem.
    
    :param image: Imagem de entrada (numpy array)
    :param salt_prob: Probabilidade de pixels brancos (sal)
    :param pepper_prob: Probabilidade de pixels pretos (pimenta)
    :return: Imagem com ruído
    """
    noisy_image = image.copy()
    total_pixels = image.size
    
    # Adiciona sal (branco)
    num_salt = int(total_pixels * salt_prob)
    salt_coords = [np.random.randint(0, i - 1, num_salt) for i in image.shape[:2]]
    noisy_image[salt_coords[0], salt_coords[1]] = 255
    
    # Adiciona pimenta (preto)
    num_pepper = int(total_pixels * pepper_prob)
    pepper_coords = [np.random.randint(0, i - 1, num_pepper) for i in image.shape[:2]]
    noisy_image[pepper_coords[0], pepper_coords[1]] = 0
    
    return noisy_image

# Exemplo de uso
if __name__ == "__main__":
    image = cv2.imread("Lenna.png", cv2.IMREAD_GRAYSCALE)  # Carregar a imagem em escala de cinza
    noisy_image = add_salt_pepper_noise(image, 0.02, 0.02)
    cv2.imwrite("/ruidosa/ruidosa.png",noisy_image)
    
    cv2.imshow("Imagem com Ruído", noisy_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
